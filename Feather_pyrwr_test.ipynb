```python
from pywr.core import Model, Input, Storage, Output, Link

# create a model (including an empty network)
model = Model()

# create two nodes: a supply, and a demand
i_orovl = Input(model, name='i_orovl')
s_orovl = Storage(model, name='s_orovl')
d_orovl = Output(model, name='d_orovl')
c_orovl = Link(model, name = 'c_orovl')

# create a connection from the inflow to the storage
i_orovl.connect(s_orovl)
s_orovl.connect(d_orovl)
s_orovl.connect(c_orovl)

# set maximum flows
d_orovl.max_flow = 15

# set storage properties
s_orovl.max_volume = 3537.577
s_orovl.initial_volume = 1900
s_orovl.min_volume = 0

# set cost (+ve) or benefit (-ve)
s_orovl.cost = 50
d_orovl.cost = 40
```

```python
import datetime
import pandas
from pywr.core import Timestepper

model.timestepper = Timestepper(
    pandas.to_datetime('2017-01-01'),  # first day
    pandas.to_datetime('2017-12-31'),  # last day
    datetime.timedelta(1)  # interval
)
```

```python
from pywr.recorders import NumpyArrayNodeRecorder

recorder = NumpyArrayNodeRecorder(model, d_orovl)
```

```python
# lets get this party started!
model.run()
```
